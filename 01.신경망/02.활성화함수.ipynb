### 활성화(Activation) 함수

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(-5, 5, 501)
```

- heaviside step function

```python
def step(x):
    y = np.zeros_like(x)
    y[x > 0] = 1
    return y
```

```python
plt.figure(figsize=(5, 3))
plt.plot(xs, step(xs))
plt.grid(), plt.title('Heaviside step function')
plt.show()
```

- sigmoid

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
plt.figure(figsize=(5, 3))
plt.plot(xs, sigmoid(xs))
plt.grid(), plt.title('Sigmoid function')
plt.show()
```

- Hyperbolic tangent

```python
plt.figure(figsize=(5, 3))
plt.plot(xs, np.tanh(xs))
plt.grid(), plt.title('Hyperbolic tangent function')
plt.show()
```

- Relu(rectified linear unit)

```python
def relu(x):
    return np.maximum(0, x)
```

```python
plt.figure(figsize=(5, 3))
plt.plot(xs, relu(xs))
plt.grid(), plt.title('Relu function')
plt.show()
```

- Leaky relu

```python
def leaky_relu(x):
    return np.maximum(0.02 * x, x)
```

```python
plt.figure(figsize=(5, 3))
plt.plot(xs, leaky_relu(xs))
plt.grid(), plt.title('Leaky Relu function')
plt.show()
```

- softplus

```python
def softplus(x):
    return np.log(1 + np.exp(x))
```

```python
plt.figure(figsize=(5, 3))
plt.plot(xs, softplus(xs))
plt.plot(xs, relu(xs))
plt.grid(), plt.title('Softplus function')
plt.show()
```

- softmax

```python
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x))
```

```python
yhat = np.array([3.0, 1.5, 0.3])
softmax(yhat)
```
